```python
np.random.seed(0)

K = 5000
fig, axs = plt.subplots(2, 3, sharey=True, tight_layout=True, figsize=(15,10))

v_mean = 279

ns = [1, 5, 10, 50, 100, 1000]
n_bins_range = [30, 40, 50, 100, 150, 500] 

for i in np.arange(len(ns)):
    n = ns[i]
    vT_all = np.zeros(K)
    for k in np.arange(K):
        v_j = np.random.default_rng().uniform(0,4*v_mean,n)
        theta_j = np.random.default_rng().uniform(0,np.pi/2,n)
        vT = np.mean(v_j*np.cos(theta_j))
        vT_all[k] = vT
    row = np.floor(np.mod(i/3,2))
    column = np.floor(np.mod(i,3))

    bins = np.linspace(0,1000,n_bins_range[i])
    #bins = np.linspace(0,1000,30)

    axs[int(row),int(column)].hist(vT_all, bins=bins)
plt.show()
```
